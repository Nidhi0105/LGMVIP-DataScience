## Image to Pencil Sketch with Python:
### We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.


### Prepared by: Nidhi Bhatia

##### Importing all the basic libararies

```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

#####  Loading our image

```python
img_org = cv2.imread("/Users/nidhibhatia/Desktop/Nidhibh.jpeg")
img = cv2.cvtColor(img_org,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(14,14))
plt.imshow(img)
plt.grid(False)
plt.title("Original Image")
plt.show()
```

We have plotted the original image with its original colour using the above code

##### Converting to GrayScale Image

```python
img_gry = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(14,14))
plt.imshow(img_gry,cmap="gray")
plt.grid(False)
plt.title("GrayScale Image")
plt.show()
```

We have turned the image into a classic black and white photo and now the image is easier to handle.

##### Negative Image - Inverting the Grayscale Image

```python
img_neg = cv2.bitwise_not(img_gry)
plt.figure(figsize=(14,14))
plt.imshow(img_neg,cmap="gray")
plt.grid(False)
plt.title("Negative Image")
plt.show()
```

Now we have inverted the grayscale image. Inversion enhances the details of the image and thus helps in studying the image with more precision and detail.

```python
img_smt = cv2.GaussianBlur(img_neg, (21, 21),sigmaX=0, sigmaY=0)
```

With this we have smoothened the image using the Gaussian function, that is helping in studying the image with more precision and detail.

##### Creating the pencil sketch

```python
img_final = cv2.divide(img_gry, 255 - img_smt, scale=255)
plt.figure(figsize=(14,14))
plt.imshow(img_final,cmap="gray")
plt.grid(False)
plt.title("Final Sketch Image")
plt.show()
```

This is our final output which is the pencil sketch of the image.

##### Displaying original image and pencil sketch image

```python
fig=plt.figure(figsize=(12,12))
#fig.add_subplot(1, 2, 2)
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Original Input Image")
plt.axis("off")
#fig.add_subplot(1, 2, 2)
plt.subplot(1,2,2)
plt.imshow(img_final,cmap="gray")
plt.title("Final Output Sketch Image")
plt.axis("off")
plt.show()
```

Here, using subplots we are displaying Input image and Output image in one frame.
